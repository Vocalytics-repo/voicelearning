```python
with open('example.txt', 'w') as file:
    file.write('example_txt')

def load_text(file_path):
    # 일반적인 한국어 인코딩 시도
    encodings = ['utf-8', 'cp949', 'euc-kr']

    for encoding in encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as f:
                text = f.read().strip()
            return text
        except UnicodeDecodeError:
            continue

    # 모든 인코딩 시도 실패 시
    print(f"Warning: Could not decode {file_path} with any encoding")
    return "[텍스트 로드 실패]"
```

```python
import os
def get_file_paths(base_directory):
    """
    디렉토리에서 PCM 파일과 텍스트 파일의 경로를 가져옵니다.
    """
    pcm_files = []
    txt_files = []

    # 모든 파일을 검색
    for root, _, files in os.walk(base_directory):
        for file in files:
            if file.endswith('.pcm'):
                pcm_path = os.path.join(root, file)
                # .pcm 확장자를 .txt로 변경하여 텍스트 파일 경로 생성
                txt_path = os.path.join(root, file.replace('.pcm', '.txt'))
                
                if os.path.exists(txt_path):
                    pcm_files.append(pcm_path)
                    txt_files.append(txt_path)
                else:
                    print("data를 찾지 못함")
    return pcm_files, txt_files
```

```python
import os

def print_directory_tree(root_dir, indent="", max_depth=3, current_depth=0, max_files=5):
    """
    디렉토리 구조를 트리 형태로 출력합니다.
    
    Parameters:
    -----------
    root_dir : str
        탐색할 루트 디렉토리 경로
    indent : str
        현재 들여쓰기 (재귀 호출용)
    max_depth : int
        탐색할 최대 깊이
    current_depth : int
        현재 탐색 깊이 (재귀 호출용)
    max_files : int
        각 디렉토리에서 보여줄 최대 파일 수
    """
    if current_depth > max_depth:
        print(f"{indent}│   ...")
        return
    
    try:
        items = os.listdir(root_dir)
    except PermissionError:
        print(f"{indent}├── [권한 거부: {root_dir}]")
        return
    except FileNotFoundError:
        print(f"{indent}├── [디렉토리 없음: {root_dir}]")
        return
    
    # 디렉토리와 파일 분리
    dirs = [item for item in items if os.path.isdir(os.path.join(root_dir, item))]
    files = [item for item in items if os.path.isfile(os.path.join(root_dir, item))]
    
    # 디렉토리 수와 파일 수 표시
    print(f"{indent}├── {os.path.basename(root_dir)}/ ({len(dirs)} 디렉토리, {len(files)} 파일)")
    
    # 파일 먼저 출력
    file_indent = indent + "│   "
    for i, file in enumerate(sorted(files)):
        if i >= max_files:
            print(f"{file_indent}├── ... 외 {len(files) - max_files}개 파일")
            break
        print(f"{file_indent}├── {file}")
    
    # 디렉토리 출력 및 재귀 탐색
    dir_indent = indent + "│   "
    for i, directory in enumerate(sorted(dirs)):
        if i == len(dirs) - 1:
            print_directory_tree(os.path.join(root_dir, directory), dir_indent, max_depth, current_depth + 1, max_files)
        else:
            print_directory_tree(os.path.join(root_dir, directory), dir_indent, max_depth, current_depth + 1, max_files)

# 시작 디렉토리 설정
start_dir = "/data"  # 루트 디렉토리 지정
print(f"디렉토리 구조: {start_dir}")
print_directory_tree(start_dir, max_depth=4, max_files=3)
```

```python
import os

# 먼저 get_file_paths 함수가 제대로 작동하는지 확인
def get_file_paths(directory):
    pcm_files = []
    txt_files = []
    
    # 디렉토리가 존재하는지 확인
    if not os.path.exists(directory):
        print(f"경고: {directory} 경로가 존재하지 않습니다.")
        return pcm_files, txt_files
    
    # 디렉토리 내 파일 검색
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.pcm'):
                pcm_files.append(os.path.join(root, file))
            elif file.endswith('.txt'):
                txt_files.append(os.path.join(root, file))
    
    return pcm_files, txt_files

for i in range(2, 19):
    print(f"___________________________________{i}번째 진행 중__________________________________________")
    
    if i < 10:
        test_directory = f"TrainData/unzipped_Speech/KsponSpeech_01/KsponSpeech_000{i}"
        file_path = f'TrainData/unzipped_Speech/KsponSpeech_01/KsponSpeech_000{i}_g2p' 
    else:
        test_directory = f"TrainData/unzipped_Speech/KsponSpeech_01/KsponSpeech_00{i}"
        file_path = f'TrainData/unzipped_Speech/KsponSpeech_01/KsponSpeech_00{i}_g2p'
    
    print(f"검색 디렉토리: {test_directory}")
    
    # 디렉토리가 없으면 생성
    if not os.path.exists(test_directory):
        print(f"경고: {test_directory} 디렉토리가 존재하지 않습니다. 건너뜁니다.")
        continue
    
    # 출력 디렉토리 생성
    if not os.path.exists(file_path):
        os.makedirs(file_path)
        print(f"디렉토리 생성됨: {file_path}")
    
    pcm_files, txt_files = get_file_paths(test_directory)
    print(f"{len(pcm_files)}개의 PCM 파일, {len(txt_files)}개의 TXT 파일을 찾았습니다.")
    
    if len(txt_files) == 0:
        print(f"경고: {test_directory}에서 텍스트 파일을 찾을 수 없습니다.")
        continue
    
    arr, arr1, arr2 = '', '', ''
    brr = []
    
    for string in txt_files:
        arr = string.split(".")
        arr1 = arr[0].split("/")
        # 경로 구성 확인 - 인덱스 오류 방지
        if len(arr1) > 4:
            arr2 = arr1[4]
            brr.append(arr2)
        else:
            print(f"경고: 예상치 못한 경로 형식: {string}")
```

```python
import os
import glob
from g2pk import G2p

# G2p 초기화
g2p = G2p()

# 기본 디렉토리 경로
base_dir = "/data/TrainData/unzipped_Speech/KsponSpeech_01"
target_base_dir = "/data/PreprocessData/KsponSpeech_01"

# 텍스트 파일 로드 함수 (다양한 인코딩 시도)
def load_text(text_file):
    """다양한 인코딩을 시도하여 텍스트 파일을 로드합니다."""
    encodings = ['utf-8', 'cp949', 'euc-kr', 'latin1']
    
    for encoding in encodings:
        try:
            with open(text_file, 'r', encoding=encoding) as f:
                return f.read().strip()
        except UnicodeDecodeError:
            continue
    
    # 모든 인코딩이 실패하면 바이너리 모드로 읽고 강제 디코딩
    try:
        with open(text_file, 'rb') as f:
            raw_data = f.read()
            # 강제로 latin1로 디코딩한 후 다시 utf-8로 인코딩
            return raw_data.decode('latin1').encode('utf-8').decode('utf-8')
    except Exception as e:
        raise ValueError(f"모든 인코딩 시도 실패: {str(e)}")

# 디렉토리 내의 PCM 및 TXT 파일 경로 가져오기
def get_file_paths(directory):
    """디렉토리 내의 모든 PCM 및 TXT 파일 경로를 반환합니다."""
    pcm_files = sorted(glob.glob(os.path.join(directory, "*.pcm")))
    txt_files = sorted(glob.glob(os.path.join(directory, "*.txt")))
    return pcm_files, txt_files

# 모든 하위 디렉토리 처리 (KsponSpeech_0001, KsponSpeech_0002, ..., KsponSpeech_0018)
for i in range(1, 19):
    print(f"___________________________________{i}번째 진행 중__________________________________________")
    if i < 10:
        source_dir = f"{base_dir}/KsponSpeech_000{i}"
        target_dir = f"{target_base_dir}/KsponSpeech_000{i}_g2p"
    else:
        source_dir = f"{base_dir}/KsponSpeech_00{i}"
        target_dir = f"{target_base_dir}/KsponSpeech_00{i}_g2p"
    
    print(f"처리할 디렉토리: {source_dir}")
    print(f"저장할 디렉토리: {target_dir}")
    
    # 소스 디렉토리가 존재하는지 확인
    if not os.path.exists(source_dir):
        print(f"소스 디렉토리가 존재하지 않습니다: {source_dir}, 건너뜁니다.")
        continue
    
    # 대상 디렉토리 생성
    os.makedirs(target_dir, exist_ok=True)
    
    # 파일 경로 가져오기
    pcm_files, txt_files = get_file_paths(source_dir)
    print(f"{len(txt_files)}개의 텍스트 파일을 처리합니다.")
    
    # 각 텍스트 파일 처리
    success_count = 0
    skip_count = 0
    error_count = 0
    
    for txt_file in txt_files:
        # 파일명 추출 (예: KsponSpeech_000001)
        base_name = os.path.basename(txt_file).split('.')[0]
        target_file = os.path.join(target_dir, f"{base_name}.txt")
        
        # 이미 처리된 파일인지 확인
        if os.path.exists(target_file):
            skip_count += 1
            # 너무 많은 로그 출력을 방지하기 위해 100개마다 출력
            if skip_count % 100 == 0:
                print(f"  {skip_count}개 파일 이미 처리됨")
            continue
        
        # 텍스트 로드 및 G2P 변환
        try:
            # 다양한 인코딩을 시도하여 텍스트 로드
            original_text = load_text(txt_file)
            
            # G2P 변환 적용
            g2p_text = g2p(original_text)
            
            # 결과 저장
            with open(target_file, 'w', encoding='utf-8') as f:
                f.write(g2p_text)
            
            success_count += 1
            # 너무 많은 로그 출력을 방지하기 위해 100개마다 출력
            if success_count % 100 == 0:
                print(f"  {success_count}개 파일 변환 완료")
        except Exception as e:
            error_count += 1
            print(f"  변환 오류 ({base_name}): {str(e)}")
            
            # 오류를 더 자세히 기록
            if "codec can't decode" in str(e):
                try:
                    # 파일 크기 확인
                    file_size = os.path.getsize(txt_file)
                    # 파일의 처음 몇 바이트 읽기
                    with open(txt_file, 'rb') as f:
                        first_bytes = f.read(20)  # 처음 20바이트
                    print(f"    파일 크기: {file_size}바이트, 처음 바이트: {first_bytes}")
                except Exception as debug_e:
                    print(f"    디버깅 시도 중 오류: {str(debug_e)}")
    
    print(f"{i}번째 디렉토리 처리 결과:")
    print(f"  성공: {success_count}개")
    print(f"  건너뜀 (이미 처리됨): {skip_count}개")
    print(f"  오류: {error_count}개")

print("모든 처리가 완료되었습니다.")
```
